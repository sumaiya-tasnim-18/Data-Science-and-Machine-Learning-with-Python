```python
import numpy as np
```

```python
# Generate 100 linearly spaced points between 50 and 60
arr = np.linspace(50, 60, 100)

print(repr(arr))
```

```python
#Reshape to 10x10 matrix
matrix = arr.reshape(10, 10)

print(matrix)
```

```python
#performing the operation, array % 5 == 0
operation = arr[arr % 5 == 0]
print(repr(operation))
```
