Data Visualization Techniques of StudentsPerformance.csv Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("StudentsPerformance.csv")
```

1. Bar Chart of Average Each scores by gender

```python
scores = ['math score', 'reading score', 'writing score']

# Create subplots: 1 row, 3 columns
fig, axes = plt.subplots(1, 3, figsize=(18,5))

for ax, score in zip(axes, scores):
    avg_scores = df.groupby('gender')[score].mean()
    avg_scores.plot(kind='bar', color=['skyblue', 'lightcoral'], ax=ax)
    ax.set_title(f'Average {score.capitalize()} by Gender')
    ax.set_xlabel('Gender')
    ax.set_ylabel(f'Average {score.capitalize()}')
    ax.set_xticklabels(avg_scores.index, rotation=0)

plt.tight_layout()
plt.show()
```

2. The count of each "parental level of education" in the dataset as a Bar Chart

```python
# Count the occurrences of each education level
education_counts = df['parental level of education'].value_counts()

# Plot bar chart
plt.figure(figsize=(10, 6))
education_counts.plot(kind='bar', color='lightgreen')
plt.title('Parental Level of Education - Count')
plt.xlabel('Parental Level of Education')
plt.ylabel('Number of Students')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

3. Count plot of Males and Females are in each race/ethnicity group

```python
sns.set(style="whitegrid")

plt.figure(figsize=(8,6))
sns.countplot(x='race/ethnicity', hue='gender', data=df, palette="Set2")
plt.title("Number of Students by Gender in Each Race/Ethnicity")
plt.xlabel("Race/Ethnicity")
plt.ylabel("Number of Students")
plt.legend(title='Gender')
plt.show()
```

4. A Heatmap showing the correlation between math, reading, and writing scores

```python
# Select only the score columns
score_data = df[['math score', 'reading score', 'writing score']]

# Compute the correlation matrix
corr_matrix = score_data.corr()

# Plot heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Between Scores')
plt.tight_layout()
plt.show()
```

5. A Pair plot showing the pairwise relationships between math, reading, and writing scores

```python
# Select score columns
score_data = df[['math score', 'reading score', 'writing score']]

# Create pair plot
sns.pairplot(score_data)
plt.suptitle('Pairwise Relationships Between Scores', y=1.02)
plt.show()
```

6. Count plots for gender, race/ethnicity, and lunch type

```python
plt.figure(figsize=(15, 4))

# Gender
plt.subplot(1, 3, 1)
sns.countplot(data=df, x='gender', hue='gender', palette='pastel', legend=False)
plt.title('Number of Students by Gender')

# Race/Ethnicity
plt.subplot(1, 3, 2)
sns.countplot(data=df, x='race/ethnicity', hue='race/ethnicity', palette='pastel', legend=False)
plt.title('Number of Students by Race/Ethnicity')

# Lunch
plt.subplot(1, 3, 3)
sns.countplot(data=df, x='lunch', hue='lunch', palette='pastel', legend=False)
plt.title('Number of Students by Lunch Type')

plt.tight_layout()
plt.show()
```

7. Scatter plot code for all three score comparisons

```python
plt.figure(figsize=(15, 4))

# Math vs Reading
plt.subplot(1, 3, 1)
sns.scatterplot(data=df, x='math score', y='reading score', hue='math score', palette='viridis', legend=False)
plt.title('Math vs Reading Score')

# Math vs Writing
plt.subplot(1, 3, 2)
sns.scatterplot(data=df, x='math score', y='writing score', hue='math score', palette='viridis', legend=False)
plt.title('Math vs Writing Score')

# Reading vs Writing
plt.subplot(1, 3, 3)
sns.scatterplot(data=df, x='reading score', y='writing score', hue='reading score', palette='viridis', legend=False)
plt.title('Reading vs Writing Score')

plt.tight_layout()
plt.show()
```

8. Box plot for scores by gender

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='gender', y='value', hue='variable', 
            data=df.melt(id_vars='gender', value_vars=['math score', 'reading score', 'writing score']),
            palette="Set2")
plt.title('Distribution of Scores by Gender')
plt.show()
```

9. Histograms for each score

```python
sns.set(style="whitegrid")

# Scores to plot
scores = ['math score', 'reading score', 'writing score']

# Create subplots: 1 row, 3 columns
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Loop through scores and axes
for ax, score in zip(axes, scores):
    sns.histplot(df[score], bins=15, kde=True, color='skyblue', ax=ax)
    ax.set_title(f'{score.capitalize()} Distribution')
    ax.set_xlabel(score.capitalize())
    ax.set_ylabel('Number of Students')

plt.tight_layout()  # Adjust spacing
plt.show()
```

10. Pie chart for test preparation course

```python
# Count the categories
prep_counts = df['test preparation course'].value_counts()

# Pie chart
plt.figure(figsize=(6,6))
plt.pie(prep_counts, labels=prep_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen'])
plt.title("Distribution of Test Preparation Course")
plt.show()
```

11. mean, median, and mode on histograms for each score

```python
scores = ['math score', 'reading score', 'writing score']

# Create subplots: 1 row, 3 columns
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for ax, score in zip(axes, scores):
    # Plot histogram
    ax.hist(df[score], bins=15, color='skyblue', edgecolor='black')
    
    # Calculate statistics
    mean_val = df[score].mean()
    median_val = df[score].median()
    mode_val = df[score].mode()[0]
    
    # Add vertical lines for mean, median, mode
    ax.axvline(mean_val, color='red', linestyle='--', label=f'Mean: {mean_val:.1f}')
    ax.axvline(median_val, color='green', linestyle='-', label=f'Median: {median_val:.1f}')
    ax.axvline(mode_val, color='orange', linestyle=':', label=f'Mode: {mode_val}')
    
    # Titles and labels
    ax.set_title(f'{score.capitalize()} Distribution')
    ax.set_xlabel(score.capitalize())
    ax.set_ylabel('Number of Students')
    ax.legend()

plt.tight_layout()
plt.show()
```
