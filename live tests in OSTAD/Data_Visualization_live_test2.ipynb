Heatmap from the given dictionary

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Given dictionary
data_dic = {
    'A': [12, 15, 45, 71],
    'B': [18, 26, 50, 80],
    'C': [32, 48, 78, 93],
    'D': [29, 35, 63, 81]
}

# Convert to DataFrame
df = pd.DataFrame(data_dic)

# Create heatmap
sns.heatmap(df, annot=True, cmap="magma")

# Show plot
plt.show()
```
